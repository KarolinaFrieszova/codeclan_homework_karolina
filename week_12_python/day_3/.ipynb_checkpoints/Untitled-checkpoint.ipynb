# Car Price Prediction

### Basic data exploration 

```python
import pandas as pd

car_sales = pd.read_csv("data/car_dekho_details.csv") # read in data
```

```python
car_sales.describe().T # overview of data
```

```python
car_sales.info()
```

```python
car_sales.isna().sum() # checking missing values
```

```python
import pandas_profiling

pandas_profiling.ProfileReport(car_sales) # further data exploration using ProfileReport()
```

```python
car_sales["make"] = car_sales["name"].str.split(" ").str[0] # extract make of the vehicle from name string
```

```python
car_sales.head(1)
```

```python
car_sales["car_age"] = 2020 - car_sales["year"] # create column with age of the vehicle
```

```python
car_sales.head(1)
```

```python
car_sales.make.unique() # too many values --> we won't be using this column further
```

```python
# create df copy with required veriables
car_sales_copy = car_sales.loc[:, ["selling_price", "km_driven", "car_age", "fuel", "seller_type", "transmission", "owner"]].copy()
```

```python
car_sales_copy.head(2)
```

```python
car_sales_copy = pd.get_dummies(car_sales_copy, drop_first = True) # create dummy veriables for categorical data

car_sales_copy.head(1)
```

```python
car_sales_copy.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,8))
sns.heatmap(car_sales_copy.corr(),annot=True);
```

We can see that the two most important predictors for selling price are 'transmission_manual' and 'car_age'.

```python
# stlit data on test and train

X = car_sales_copy.drop(columns = "selling_price")
y = car_sales_copy["selling_price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# building model

from sklearn.linear_model import LinearRegression

model = LinearRegression() # define our model object

model.fit(X_train, y_train) # fit to the training data
```

```python
model.score(X_train, y_train) # R-squared value
```

```python
model.score(X_test, y_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
pd.DataFrame(
  {
  "Variable" : X.columns.values,
  "Coefficient" : model.coef_
  }
)
```

```python
import statsmodels.api as sm

X_i = sm.add_constant(X)
X_i.head()
```

```python
sm_model = sm.OLS(y, X_i).fit()
print(sm_model.summary())
```

```python
#first calculate the residuals

residuals = sm_model.fittedvalues - y
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

sns.residplot(x = sm_model.fittedvalues, y = residuals,
                          lowess=True,
                          scatter_kws={'alpha': 0.5},
                          line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
plt.xlabel("Fitted value")
plt.ylabel("Residual")
```

```python
import scipy.stats as stats

sm.qqplot(residuals, dist=stats.t, fit=True, line='45');
```

 Our data have more extreme values than would be expected.
